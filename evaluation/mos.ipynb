```python
# =========================
# libraries
# =========================
import os
import librosa
import soundfile as sf
import numpy as np
from IPython.display import Audio, display
```

```python

# =========================
# SETTINGS
# =========================
input_dir = "/content/drive/MyDrive/MusicGen_Project/Data/Apollo_Output_From_AudioSR_To_96k"
output_dir = "/content/drive/MyDrive/MusicGen_Project/Data/Normalized/Apollo_From_AudioSR_Output_Normalized"
target_rms = 0.1  # target loudness

os.makedirs(output_dir, exist_ok=True)

# =========================
# RMS normalization function
# =========================
def normalize_rms(audio, target_rms=0.1):
    rms = np.sqrt(np.mean(audio**2))
    if rms > 0:
        audio = audio * (target_rms / rms)
    return audio

# =========================
# Process all files
# =========================
for file in os.listdir(input_dir):
    if file.endswith(".wav"):
        path = os.path.join(input_dir, file)
        audio, sr = librosa.load(path, sr=None, mono=True)

        audio_norm = normalize_rms(audio, target_rms)

        out_path = os.path.join(output_dir, file)
        sf.write(out_path, audio_norm, sr)

print("Normalization finished.")
```

```python
# =========================
# Folders
# =========================
raw_dir = "/content/drive/MyDrive/MusicGen_Project/Data/Normalized/MusicGen_Output_Normalized"
dsp_dir = "/content/drive/MyDrive/MusicGen_Project/Data/Normalized/DSP_Output_Normalized"
flash_dir = "/content/drive/MyDrive/MusicGen_Project/Data/Normalized/FlashSR_Output_Normalized"
audiosr_dir = "/content/drive/MyDrive/MusicGen_Project/Data/Normalized/AudioSR_Output_Normalized"
apollo_dir = "/content/drive/MyDrive/MusicGen_Project/Data/Normalized/Apollo_From_AudioSR_Output_Normalized"

# =========================
# Select 2 random songs per category
# =========================
selected_files = [
    "ambient_02.wav", "ambient_04.wav",
    "piano_02.wav", "piano_05.wav",
    "guitar_02.wav", "guitar_05.wav",
    "drums_02.wav", "drums_05.wav",
    "electronic_02.wav","electronic_05.wav",
    "orchestral_02.wav", "orchestral_03.wav"
]

# =========================
# Play all versions for MOS
# =========================
for file in selected_files:
    print("\n======================================")
    print("Sample:", file)
    print("======================================")

    paths = {
        "A": os.path.join(raw_dir, file),
        "B": os.path.join(dsp_dir, file.replace(".wav", "_DSP.wav")),
        "C": os.path.join(flash_dir, file),
        "D": os.path.join(audiosr_dir, file),
        "E": os.path.join(apollo_dir, file)
    }

    for label, path in paths.items():
        if os.path.exists(path):
            print(f"Version {label}")
            display(Audio(path))
        else:
            print(f"Missing file for Version {label}: {path}")
```
