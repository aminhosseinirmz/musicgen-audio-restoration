```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# -------------------------
# 1) Choose sample
# -------------------------
sample_id = "piano_05"

# -------------------------
# 2) Paths
# -------------------------
paths = {
    "Unprocessed": "/content/drive/MyDrive/MusicGen_Project/Data/Normalized/MusicGen_Output_Normalized",
    "DSP": "/content/drive/MyDrive/MusicGen_Project/Data/Normalized/DSP_Output_Normalized",
    "FlashSR": "/content/drive/MyDrive/MusicGen_Project/Data/Normalized/FlashSR_Output_Normalized",
    "AudioSR": "/content/drive/MyDrive/MusicGen_Project/Data/Normalized/AudioSR_Output_Normalized",
    "AudioSR+Apollo": "/content/drive/MyDrive/MusicGen_Project/Data/Normalized/Apollo_From_AudioSR_Output_Normalized",
}

# -------------------------
# 3) Filenames
# -------------------------
file_map = {
    "Unprocessed": f"{sample_id}.wav",
    "DSP": f"{sample_id}_DSP.wav",
    "FlashSR": f"{sample_id}.wav",
    "AudioSR": f"{sample_id}.wav",
    "AudioSR+Apollo": f"{sample_id}.wav",
}

# -------------------------
# 4) Settings
# -------------------------
N_FFT = 2048
HOP_LENGTH = 512
SHOW_SECONDS = 2.5   # crop time so panels are shorter
TOP_DB = 80          # control dynamic range

# -------------------------
# 5) Figure
# -------------------------
num_methods = len(paths)
fig, axes = plt.subplots(1, num_methods, figsize=(2.6 * num_methods, 2.2), dpi=200)

if num_methods == 1:
    axes = [axes]

# -------------------------
# 6) Plot
# -------------------------
for ax, (method, folder) in zip(axes, paths.items()):
    file_path = os.path.join(folder, file_map[method])

    if not os.path.exists(file_path):
        ax.text(0.5, 0.5, "Not found", ha="center", va="center", fontsize=8)
        ax.axis("off")
        continue

    # load audio
    y, sr = librosa.load(file_path, sr=None, mono=True)

    # keep only first few seconds for compact figure
    y = y[:int(SHOW_SECONDS * sr)]

    # STFT
    D = librosa.stft(y, n_fft=N_FFT, hop_length=HOP_LENGTH)
    S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max, top_db=TOP_DB)

    # plot with LOG frequency axis
    librosa.display.specshow(
        S_db,
        sr=sr,
        hop_length=HOP_LENGTH,
        x_axis=None,
        y_axis="log",
        cmap="magma",
        ax=ax
    )

    ax.set_title(method, fontsize=9, pad=4)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlabel("")
    ax.set_ylabel("")

    for spine in ax.spines.values():
        spine.set_visible(True)
        spine.set_linewidth(0.8)
        spine.set_color("black")

plt.subplots_adjust(wspace=0.03, hspace=0)
plt.show()
```
